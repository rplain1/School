# Setup

```python
ndata = M
m, n = ndata.shape
C = (ndata.T @ ndata) / m
```

#### PCA

```python
d = 4
v, _, _ = np.linalg.svd(C)
v = v[:, :d]
```

```python
pdata = ndata.dot(v)
```

## EM

```python
K = 2
seed = 12345

np.random.seed(seed)

pi = np.random.random(K)
pi = pi/np.sum(pi)

np.random.seed(seed)

mu = np.random.randn(K, d)
mu_old = mu.copy()

sigma = []
for ii in range(K):
    dummy = np.random.randn(d, d)
    sigma.append(dummy @ dummy.T)
    
tau = np.full((m, K), fill_value=0.)

plt.ion()

maxIter = 100
tol = 1e-3

likelihoods = []

for ii in range(maxIter):
    print("--- ", ii)
    for kk in range(K):
        tau[:, kk] = pi[kk] * mvn.pdf(pdata, mu[kk], sigma[kk])

            
    sum_tau = np.sum(tau, axis=1)
    
    likelihoods.append(np.sum(np.log(sum_tau)))
    
    sum_tau.shape = (m, 1)
    tau = np.divide(tau, np.tile(sum_tau, (1, K)))
    
    
    
    for kk in range(K):
        # update prior
        pi[kk] = np.sum(tau[:, kk])/m
        
        # update component mean
        mu[kk] = pdata.T @ tau[:,kk] / np.sum(tau[:,kk], axis = 0)
        
        # update cov matrix
        dummy = pdata - np.tile(mu[kk], (m,1)) # X-mu
        sigma[kk] = dummy.T @ np.diag(tau[:,kk]) @ dummy / np.sum(tau[:,kk], axis = 0)
    
    
    
    
    
    print('-----iteration---',ii)    
    plt.scatter(pdata[:,0], pdata[:,1], c= tau[:, 1])
    plt.axis('scaled')
    plt.draw()
    plt.pause(0.1)
    if np.linalg.norm(mu-mu_old) < tol:
        print('training coverged')
        break
    mu_old = mu.copy()
    if ii==99:
        print('max iteration reached')
        break
```
